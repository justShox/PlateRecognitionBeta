```python
import torch

# Load your custom YOLOv5 model
try:
    model = torch.hub.load('ultralytics/yolov5', 'custom', path='./models/plate_detection.pt')
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")

# Set the model to evaluation mode
model.eval()
```

```python
import cv2

# Load an image
image_path = '/content/yolov5/test_images/license1.jpg'
image = cv2.imread(image_path)

# Perform inference
results = model(image)

# Parse results
detections = results.xyxy[0].numpy()  # Detections in format [x1, y1, x2, y2, confidence, class]
```

```python
import easyocr

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])  # Specify language(s)

# Loop through detections
for detection in detections:
    x1, y1, x2, y2, conf, cls = detection
    if cls == 0:  # Assuming class 0 is the license plate class
        # Crop the license plate region
        license_plate = image[int(y1):int(y2), int(x1):int(x2)]

        # Use EasyOCR to extract text
        ocr_results = reader.readtext(license_plate)

        # Print OCR results
        for result in ocr_results:
            text = result[1]
            print(f"Detected Text: {text}")
```

```python
# Draw bounding boxes and text on the image
for detection in detections:
    x1, y1, x2, y2, conf, cls = detection
    if cls == 0:
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(image, text, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Display the image
cv2.imshow("License Plate Detection", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the image
cv2.imwrite("output.jpg", image)
```
